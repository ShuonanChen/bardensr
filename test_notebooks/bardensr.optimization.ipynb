# test cg

Without preconditioner

```python
n=200

npr.seed(0)
A=npr.randn(n,n)
A=A@A.T

x_tru=npr.randn(n)
b=A@x_tru

A=tf.identity(A)
x_tru=tf.identity(x_tru)
b=tf.identity(b)

cgs=bardensr.optimization.cg_tf.cg(
    lambda x: tf.linalg.matvec(A,x),
    b,
    50,
)
est_x=cgs.x


print('variance unexplained',(tf.reduce_sum(cgs.r**2).numpy()/tf.reduce_sum(b**2)).numpy())
print('resid',tf.reduce_sum(cgs.r**2).numpy())
print('resid',tf.reduce_sum((tf.linalg.matvec(A,est_x)-b)**2).numpy())
print('iters',cgs.i)

plt.scatter(est_x.numpy(),x_tru.numpy())
plt.plot([-1,1],[-1,1],'r-')
```

with preconditioner

```python

precond=np.linalg.inv(A.numpy())
precond=tf.identity(precond)

cgs=bardensr.optimization.cg_tf.cg(
    lambda x: tf.linalg.matvec(A,x),
    b,
    5,
    preconditioner=lambda x: tf.linalg.matvec(precond,x)
)

est_x=cgs.x

print('variance unexplained',(tf.reduce_sum(cgs.r**2).numpy()/tf.reduce_sum(b**2)).numpy())
print('resid',tf.reduce_sum(cgs.r**2).numpy())
print('resid',tf.reduce_sum((tf.linalg.matvec(A,est_x)-b)**2).numpy())
print('iters',cgs.i)

plt.scatter(est_x.numpy(),x_tru.numpy())
plt.plot([-1,1],[-1,1],'r-')
```

```python
npr.seed(0)
x_guess=tf.identity(x_tru.numpy()+npr.randn(n)*.1)
cgs=bardensr.optimization.cg_tf.cg(
    lambda x: tf.linalg.matvec(A,x),
    b,
    50,
    x0=x_guess
)

est_x=cgs.x


print('variance unexplained',(tf.reduce_sum(cgs.r**2).numpy()/tf.reduce_sum(b**2)).numpy())
print('resid',tf.reduce_sum(cgs.r**2).numpy())
print('resid',tf.reduce_sum((tf.linalg.matvec(A,est_x)-b)**2).numpy())
print('iters',cgs.i)


plt.scatter(est_x.numpy(),x_tru.numpy())
plt.plot([-1,1],[-1,1],'r-')
```

### what if we enforce positivity?

```python
'''
try to solve 

    min   ||blur(x)-Abl||^2 - lam*||x||_1
    s.t.  x >=0
          
'''


@tf.function
def loss(x,target,lam):
    blurx   = bardensr.blur_kernels.gaussian_filter_2d(x,tf.identity([5,5]))
    return tf.reduce_mean((blurx-target)**2) + lam*tf.reduce_sum(x)
    
    
@tf.function
def setup_ap_with_newton(x,target,lam,maxiter):
    sd=bardensr.optimization.hvp_tf.inexact_newton_solve_with_bounds(
        lambda x:loss(x,target,lam),
        x,
        maxiter=maxiter,
        lo=0
    )
    
    return bardensr.optimization.hvp_tf.prepare_for_armijo_linesearch(
        lambda x: loss(x,target,lam),
        x,
        lo=0.0,
        search_direction=sd
    )

@tf.function
def setup_ap_without_newton(x,target,lam):
    return bardensr.optimization.hvp_tf.prepare_for_armijo_linesearch(
        lambda x: loss(x,target,lam),
        x,
        lo=0.0,
    )

# # initial condition
# x=tf.identity(np.zeros(A.shape)) # from zero, not so much...
# target=Abl_tf
# lam=tf.cast(0,x.dtype)
# ap=setup_ap_without_newton(x,target,lam)
```

```python


# initial condition
x=tf.identity(np.zeros(A.shape)) # from zero, not so much...

# lambda
lam=tf.cast(0,x.dtype)

# target
target = Abl_tf

# loop
losses=[loss(x,target,lam).numpy()]
tq=tqdm.notebook.trange(20)
for i in tq:
    ap=setup_ap_with_newton(x,target,lam,tf.identity(20))
    x,newloss,armijo,i=bardensr.optimization.hvp_tf.armijo_linesearch(
        lambda x: loss(x,target,lam),
        ap,
        maxiter=10
    )
    losses.append(loss(x,target,lam).numpy())
    tq.set_description(f'{losses[-1]:.4e}')
    

tq=tqdm.notebook.trange(500)
for i in tq:
    ap=setup_ap_without_newton(x,target,lam)
    x,newloss,armijo,i=bardensr.optimization.hvp_tf.armijo_linesearch(
        lambda x: loss(x,target,lam),
        ap,
        maxiter=10
    )
    losses.append(loss(x,target,lam).numpy())
    tq.set_description(f'{losses[-1]:.4e}')
        
        
print(losses[-1])
```

```python
with bardensr.plotting.AnimAcross() as a:
    a('data')
    a.cb(plt.imshow(Abl))
    
    a('estimated x')
    a.cb(plt.imshow(x))
    
    a('reconstruction')
    recon=bardensr.blur_kernels.gaussian_filter_2d(x,(5,5)).numpy()
    a.cb(plt.imshow(recon))
    
    a('losses')
    plt.plot(losses)
    plt.gca().set_yscale('log')
```

```python
plt.plot(recon.ravel(),Abl.ravel(),'.',alpha=.1)
plt.xlabel("reconstruction")
plt.ylabel("original")
```

### we could also use an LP, but interior point methods are hard

```python
'''
try to solve 

    max   sum blur(x)*Abl
    s.t.  x >=0
          blur(x) < Abl+eps
          
via interior point method
'''

Abl_tf_plus_noise=Abl_tf + tf.identity(npr.rand(*Abl_tf.shape))*.0001+.0001

# define barrier loss
@tf.function
def loss(x,eps):
    eps=tf.cast(eps,x.dtype)
    blurx   = bardensr.blur_kernels.gaussian_filter_2d(x,tf.identity([5,5]))
    matchy  = tf.reduce_mean(x*Abl_tf)
    barrier = tf.reduce_mean(tf.math.log(Abl_tf_plus_noise-blurx))
    total=-matchy-eps*barrier
    return total,matchy,barrier
    
@tf.function
def setup_ap_with_newton(x,eps,maxiter):
    sd=bardensr.optimization.hvp_tf.inexact_newton_solve_with_bounds(
        lambda x:loss(x,eps)[0],
        x,
        maxiter=maxiter,
        lo=0
    )
    
    return bardensr.optimization.hvp_tf.prepare_for_armijo_linesearch(
        lambda x: loss(x,eps)[0],
        x,
        lo=0.0,
        search_direction=sd
    )

@tf.function
def setup_ap_without_newton(x,eps):
    sd=None
    
    return bardensr.optimization.hvp_tf.prepare_for_armijo_linesearch(
        lambda x: loss(x,eps)[0],
        x,
        lo=0.0,
        search_direction=sd
    )
    
```

```python

# initial condition
x=tf.identity(np.zeros(A.shape)) # from zero, not so much...
# x=tf.identity(A) # <-- if we provide good inital condition, it does pretty well
    
# set initial epsilon 
eps=tf.identity(.001) # yeesh.  how to pick this?

# loop
losses=[-loss(x,eps)[1].numpy()]
tq=tqdm.notebook.trange(20)
for i in tq:
    eps=eps*.99  # ugh, how to pick this?
    ap=setup_ap_with_newton(x,eps,maxiter=tf.identity(20))
#     ap=setup_ap_without_newton(x,eps)
    if tf.reduce_sum(ap.search_direction**2)>0:
        x,newloss,armijo,i=bardensr.optimization.hvp_tf.armijo_linesearch(
            lambda x: loss(x,eps)[0],
            ap,
            maxiter=10
        )
        losses.append(-loss(x,eps)[1].numpy())
        tq.set_description(f'{losses[-1]:.5f}')
        
        
print(losses[-1])
```

```python
with bardensr.plotting.AnimAcross() as a:
    a('data')
    a.cb(plt.imshow(Abl))
    
    a('estimated x')
    a.cb(plt.imshow(x))
    
    a('reconstruction')
    recon=bardensr.blur_kernels.gaussian_filter_2d(x,(5,5)).numpy()
    a.cb(plt.imshow(recon))
    
    a('losses')
    plt.plot(losses)
```

```python
bench = bardensr.benchmarks.load_h5py('/home/jovyan/work/Downloads/smallbench.hdf5')

A = tf.convert_to_tensor(bench.X.copy())
Abl_tf = A


# define barrier loss


@tf.function
def loss(x,target,lam,blur):
    blurx   = bardensr.blur_kernels.gaussian_filter_3d(x,blur)
    return .5*tf.reduce_sum((blurx-target)**2)+tf.reduce_sum(x)*lam

@tf.function
def setup_ap_without_newton(x,target,lam,blur):
    return bardensr.optimization.hvp_tf.prepare_for_armijo_linesearch(
        lambda x: loss(x,target,lam,blur),
        x,
        lo=0.0,
    )
   
```

```python
R,C,M0,M1,M2 = A.shape
deblurred_x=np.zeros(A.shape)
```
